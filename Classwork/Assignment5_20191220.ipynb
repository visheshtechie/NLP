```
#Importing Libraries and Data
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from gensim.models import Word2Vec
nltk.download('stopwords')
nltk.download('punkt')
#Setting english stopwords
stop_words = set(stopwords.words('english'))
#Reading Data
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
```

```
#Creating a dataframe with only reviewText column
df=data.filter(['reviewText'])
```

```
df.head()
```

```
#Function to lowercase, remove punctuation,remove stopwords and tokenize the string
def tokenize(text):
  text=text.lower()
  seps='''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  for sep in seps:
    text=text.replace(sep,"")
  temp=word_tokenize(text)
  tokens = [w for w in temp if not w in stop_words] 
  return tokens
```

```
#Creating a column in dataframe that displays tokenized form of reviewText column
df['Tokens']=df.reviewText.apply(tokenize)
```

```
df.head()
```

```
#Creating a list that will hold Tokens from the whole reviewText data
MyWords = []
for i in range(0,df.shape[0]):
  for j in df.Tokens[i]:
    MyWords.append(j)
    
#Number of tokens in reviewText
len(MyWords)
```

```
#Function to keep only unique elements in a list
def uWords(list1):
  uList=list(set(list1))
  return uList
```

```
#Creating a list that will hold only unique tokens of reviewText
Unique = uWords(MyWords)

#Checking the number of unique words
len(Unique)
```

```
df['Unique_Tokens']=df.Tokens.apply(uWords)
df.head()
```

```
#Creating Word2Vec Training data
wordlist = [[w] for w in df.Unique_Tokens[0]]
```

```
for x in wordlist:
  print (x)
```

```
#Creating Word2Vec Model
model=Word2Vec(wordlist, size=100, window=5, min_count=1, workers=4)
```

```
#Checking the word similarities
for i in wordlist:
  for j in wordlist:
    if j>i:
      print(i[0],' :',j[0],model.similarity(i[0],j[0]))
    else:
      continue
```
